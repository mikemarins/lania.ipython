```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys, os, re, time
import numpy as np
import cv2
```

```python
pic = os.path.join('.','ZuzanaB.jpg')

img=cv2.imread(pic)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
caras_cascada=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

caras = caras_cascada.detectMultiScale(gray, 1.3, 4)

print "Encontramos {0} faces!".format(len(caras))
```

```python
i=img.copy()
for (x,y,w,h) in caras:
    cv2.rectangle(i, (x,y), (x+w,y+h), (0,255,0), 2)
plt.imshow(i)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys, os, re, time
import numpy as np
import cv2
import android
droide=android.Android()
droide.cameraCapturePicture('/sdcard/DCIM/CAMERA/carota.jpg')
!adb pull /sdcard/DCIM/CAMERA/carota.jpg
pic = os.path.join('.','carota.jpg')

img=cv2.imread(pic)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
caras_cascada=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
caras = caras_cascada.detectMultiScale(gray, 1.1, 2)
print "Encontramos {0} faces!".format(len(caras))
i=img.copy()
for (x,y,w,h) in caras:
    cv2.rectangle(i, (x,y), (x+w,y+h), (0,255,0), 2)
plt.imshow(i)
```
